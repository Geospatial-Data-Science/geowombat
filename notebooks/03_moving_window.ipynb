### Moving window operations

```python
import geowombat as gwb
```

---

```python
fn = '../data/test.tif'
```

```python
with gwb.open(fn) as src:
    garray = src.read(bands2open=-1)
```

---

```python
mean = garray.moving(5).mean()
```
